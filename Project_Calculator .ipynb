```python
import tkinter as tk
import math

# Function to update the input field
def on_button_click(value):
    entry_field.insert(tk.END, value)

# Function to evaluate the expression
def calculate():
    try:
        expression = entry_field.get().replace("^", "**")  # Replacing ^ with ** for exponentiation
        result = eval(expression)
        entry_field.delete(0, tk.END)
        entry_field.insert(tk.END, str(result))
    except Exception:
        entry_field.delete(0, tk.END)
        entry_field.insert(tk.END, "Error")

# Function to clear the input field
def clear():
    entry_field.delete(0, tk.END)

# Function to remove last character (Backspace)
def backspace():
    current_text = entry_field.get()
    entry_field.delete(0, tk.END)
    entry_field.insert(tk.END, current_text[:-1])

# Function to calculate square root
def square_root():
    try:
        number = float(entry_field.get())
        result = math.sqrt(number)
        entry_field.delete(0, tk.END)
        entry_field.insert(tk.END, str(result))
    except Exception:
        entry_field.delete(0, tk.END)
        entry_field.insert(tk.END, "Error")

# Creating main window
root = tk.Tk()
root.title("Advanced Calculator")
root.geometry("350x500")
root.resizable(False, False)

# Entry field for user input
entry_field = tk.Entry(root, font=("Arial", 18), justify="right", bd=10, relief=tk.GROOVE)
entry_field.pack(fill="both", padx=10, pady=10)

# Button layout
buttons = [
    ("7", "8", "9", "/"),
    ("4", "5", "6", "*"),
    ("1", "2", "3", "-"),
    ("C", "0", "=", "+"),
    ("⌫", "√", "^", "%")  # Backspace, Square Root, Power, Modulus
]

# Creating buttons and placing them in the window
frame = tk.Frame(root)
frame.pack()

for row in buttons:
    row_frame = tk.Frame(frame)
    row_frame.pack(fill="both", expand=True)
    for button in row:
        if button == "=":
            action = calculate
        elif button == "C":
            action = clear
        elif button == "⌫":
            action = backspace
        elif button == "√":
            action = square_root
        else:
            action = lambda x=button: on_button_click(x)
        
        tk.Button(row_frame, text=button, font=("Arial", 14), width=5, height=2, command=action).pack(side="left", expand=True)

# Runing the GUI loop
root.mainloop()
```
